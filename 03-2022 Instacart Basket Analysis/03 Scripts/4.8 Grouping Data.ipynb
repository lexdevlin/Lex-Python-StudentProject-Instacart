```python
# Q6
# Create a column denoting a user's average spend
ords_prods_merge['avg_spend'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge[['user_id', 'prices', 'avg_spend']]
```

```python
# Create spending flag
ords_prods_merge.loc[ords_prods_merge['avg_spend'] >= 10, 'spending_flag'] = 'High spender'
ords_prods_merge.loc[ords_prods_merge['avg_spend'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge[['user_id', 'avg_spend', 'spending_flag']]
```

```python
# Q7
# Create a column denoting a user's average days since prior order
ords_prods_merge['avg_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.mean)
```

```python
ords_prods_merge[['user_id', 'days_since_prior_order', 'avg_days_since_prior_order']].head(60)
```

```python
# Create frequency flag
ords_prods_merge.loc[ords_prods_merge['avg_days_since_prior_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['avg_days_since_prior_order'] > 10) & (ords_prods_merge['avg_days_since_prior_order'] <= 20), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['avg_days_since_prior_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge[['user_id', 'avg_days_since_prior_order', 'frequency_flag']].head
```
