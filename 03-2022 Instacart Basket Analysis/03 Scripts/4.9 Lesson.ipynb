# 05. Sampling Data

```python
# Setting seed and producing boolean list for 70/30 split
np.random.seed(4)
dev = np.random.rand(len(df)) <= 0.7
```

```python
# 70/30 split of df
big = df[dev]
small = df[~dev]
```

```python
len(df)
```

```python
len(big) + len(small)
```

```python
# Create 30% subset of dataframe, with only necessary columns
df_2 = small[['orders_day_of_week', 'prices']]
```

# 06. Line Charts

```python
# Create line plot with the subset
line_2 = sns.lineplot(data = df_2, x = 'orders_day_of_week', y = 'prices')
```
