# 03. Deriving Variables

User-Defined Functions

```python
# Deriving price_label with if/then statements
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High-range product'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Values seem out of place.
df['price_range'].value_counts()
```

```python
# Q2 - For-loop designating values for busiest_days
result_1 = []

for value in df_ords_prods["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_1.append("Busiest days")
  elif value == 3 or value == 4:
    result_1.append("Slowest days")
  else:
    result_1.append("Regularly busy")
```

```python
# For-loop designating busiest_period_of_day values

result_2 = []

for Z in df_ords_prods['order_hour_of_day']:
    if Z in (10, 11, 14, 15, 13, 12, 16, 9):
        result_2.append("Most orders")
    elif Z in (17, 8, 18, 19, 20, 7, 21, 22):
        result_2.append("Average orders")
    elif Z in (23, 6, 0, 1, 5, 2, 4, 3):
        result_2.append("Fewest orders")
    else:
        result_2.append("Unknown")
```

```python
# Creating busiest_period_of_day column
df_ords_prods['busiest_period_of_day'] = result_2
```
